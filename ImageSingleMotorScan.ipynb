```python
from PIL import Image
import pandas as pd
from glob import glob
import cv2
import numpy as np
```

Glob the file paths into a List (str)

```python
png_paths = glob('/Users/josuecastellanos/Documents/ALS/Image Single Motor Scan 000596 Images/*.Png')
png_paths
```

Read the .png data into a List

```python
imgs = []
for  i in png_paths:
    img_cv2 = cv2.imread(i)
    img_cv2_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
    img_cv2_rgb_intensity_values = img_cv2_rgb/255
    # imgs.append(img_cv2_rgb_intensity_values)

    flatten_real_intensity_values = img_cv2_rgb_intensity_values.flatten()
    series_flatten_values = pd.Series(flatten_real_intensity_values)
    imgs.append(series_flatten_values)

imgs    
```

```python
for i in png_paths:
    image = Image.open(i)
    image = image.convert('RGB')
    width, height = image.size
    
```

```python

data = []
for y in range(height):
    for x in range(width):
        r, b, g = image.getpixel((x,y))
        hex_value = f'#{r:02x}{g:02x}{b:02x}'
        data.append(((x,y), hex_value))

data
# df = pd.DataFrame(data, columns=['Coordinates', 'HexValue'])
# df.set_index('Coordinates', inplace=True)
# df
```
